```python
df_stats_bert = df_metrics_bert.describe().loc[['std']]
df_stats_bert
```

```python
import pandas as pd

# Data hasil cross-validation
metrics_bilstm = [
    {"Fold": 1, "Train Accuracy": 0.9409, "Val Accuracy": 0.8127, "Precision": 0.8230, "Recall": 0.8127, "F1 Score": 0.8153},
    {"Fold": 2, "Train Accuracy": 0.9582, "Val Accuracy": 0.8394, "Precision": 0.8455, "Recall": 0.8394, "F1 Score": 0.8386},
    {"Fold": 3, "Train Accuracy": 0.9294, "Val Accuracy": 0.8511, "Precision": 0.8580, "Recall": 0.8511, "F1 Score": 0.8482},
    {"Fold": 4, "Train Accuracy": 0.9433, "Val Accuracy": 0.8411, "Precision": 0.8461, "Recall": 0.8411, "F1 Score": 0.8389},
    {"Fold": 5, "Train Accuracy": 0.9411, "Val Accuracy": 0.8502, "Precision": 0.8546, "Recall": 0.8502, "F1 Score": 0.8493},
]

# Buat DataFrame
df_metrics_bilstm = pd.DataFrame(metrics_bilstm)

# Hitung rata-rata untuk tiap metrik (tanpa kolom "Fold")
mean_metrics_bilstm = df_metrics_bilstm.drop(columns=["Fold"]).mean()
mean_metrics_bilstm["Fold"] = "Mean"

# Tambahkan mean ke DataFrame
df_metrics_bilstm = pd.concat([df_metrics_bilstm, pd.DataFrame([mean_metrics_bilstm])], ignore_index=True)

# Tampilkan hasil
df_metrics_bilstm
```

```python
# Ubah index menjadi Fold 1, Fold 2, ..., Mean
df_metrics_bilstm.index = [f"Fold {i+1}" for i in range(len(df_metrics_bilstm)-1)] + ["Mean"]

# Drop kolom 'Fold'
df_metrics_bilstm = df_metrics_bilstm.drop(columns=['Fold'])

df_metrics_bilstm
```

```python
df_stats_bilstm = df_metrics_bilstm.describe().loc[['std']]
df_stats_bilstm
```

```python
from scipy.stats import ttest_rel
import pandas as pd

# Buang baris "Mean" berdasarkan index
df_bert = df_metrics_bert.drop(index="Mean")
df_bilstm = df_metrics_bilstm.drop(index="Mean")

# Pastikan data dalam bentuk float
df_bert = df_bert.astype(float)
df_bilstm = df_bilstm.astype(float)

# Kolom metrik yang ingin diuji
metrics = df_bert.columns  # atau tuliskan manual: ["Train Accuracy", "Val Accuracy", ...]

# Lakukan t-test
results = {}
for metric in metrics:
    t_stat, p_value = ttest_rel(df_bert[metric], df_bilstm[metric])
    results[metric] = {"t-stat": t_stat, "p-value": p_value}

# Tampilkan hasil
t_test_results = pd.DataFrame(results).T
t_test_results
```
